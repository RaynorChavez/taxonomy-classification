```python
import pandas as pd

df = pd.read_csv('full_captions_taxons_dataset.csv').rename({'new_captions': 'captions'}, axis=1)
df
```

```python
from ast import literal_eval

df['captions'] = df.apply(lambda x: [y.strip("'") for y in literal_eval(x['captions'])], axis=1)
```

```python
test_data = df[df.duplicated('species')].sample(frac=0.75, random_state=133)
test_data
```

```python
train_data = df.drop(test_data.index, axis=0)
train_data
```

```python
train_data.to_json('train_dataset.json', orient='records')
```

```python
test_data.to_json('test_dataset.json', orient='records')
```
